```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
}
```

```python
class Website:
    def __init__(self, url):
        self.url = url
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')
        self.title = soup.title.string if soup.title else "No title found"
        
        for irrelevant in soup.body(["script", "style", "img", "input", "link", "ul", "li", "a"]):
            irrelevant.decompose()
        
        self.text = soup.body.get_text(separator="\n", strip=True)
```

```python
web = Website("https://pl.wikipedia.org/wiki/Du%C5%BCy_model_j%C4%99zykowy")
print("Title:", web.title)
print("Text:\n", web.text)
```

```python
system_prompt = (
    "You are an assistant that analyzes the contents of a website "
    "and provides a short summary, ignoring text that might be navigation related. "
    "Respond in markdown."
)
```

```python
def user_prompt_for(website):
    user_prompt = f"You are looking at a website titled {website.title}.\n"
    user_prompt += (
        "The contents of this website is as follows; please provide a short summary of this website in markdown. "
        "If it includes news or announcements, then summarize these too.\n\n"
    )
    user_prompt += website.text
    return user_prompt
```

```python
def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_for(website)}
    ]
```

```python
MODEL = 'llama3.2'

# Funkcja podsumowująca stronę
def summarize(url):
    website = Website(url)
    response = ollama.chat(
        model=MODEL,
        messages=messages_for(website)
    )
    return response['message']['content']
```

```python
summarize("https://wp.pl")
```

```python
def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))
```

```python
display_summary("https://wp.pl")
```
